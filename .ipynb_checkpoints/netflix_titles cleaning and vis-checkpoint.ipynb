```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("netflix_titles.csv")
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.drop(labels= "show_id", axis = 1 , inplace= True)
```

```python
data.head()
```

```python
data['date_added'] = pd.to_datetime(data['date_added'])
data["year_added"] = data["date_added"].dt.year
data["month_added"]= data["date_added"].dt.month
data['Day_Name'] = data['date_added'].dt.day_name()
```

```python
data.head()
```

```python
data["type"].value_counts()
```

```python
sns.countplot(x = data.type)
```

```python
Movies = data[data["type"] == "Movie"]
```

```python
Movies.head()
```

```python
Movies.shape
```

```python
Movies["director"].value_counts()
```

```python
Movies["country"].value_counts()
```

```python
values = Movies.country.value_counts()[:10].tolist()
labels = Movies.country.value_counts()[:10].index.tolist()
```

```python
plt.figure(figsize=(10, 10))
plt.pie(values , labels=labels , autopct='%1.0f%%'  )
plt.title('Top 10 countries with movies and tv shows') ;
```

```python
plt.figure(figsize=(12, 10))
sns.barplot( labels , values )
plt.title('Top 10 countries with movies'); 
plt.show()
```

```python
Movies["release_year"].value_counts()
```

```python
sns.countplot(x =Movies.release_year, order = Movies["release_year"].value_counts().index[:14])
plt.xticks(rotation = 90)
plt.show()
```

```python
Movies.rating.value_counts()
```

```python
plt.figure(figsize=(14 , 10 ))
sns.countplot(data = Movies , x =Movies.rating ,order=Movies['rating'].value_counts().index[0:14] , palette= 'Set1' );
```

```python
Movies.sample()
```

```python
Movies["duration"] = Movies["duration"].str.replace("min", "")
Movies["duration"]
```

```python
Movies['duration'].value_counts()
```

```python
Movies['duration'] = pd.to_numeric(Movies['duration'], errors='coerce')
Movies['duration']
```

```python
plt.figure(figsize=(7, 5))
sns.set(style='darkgrid')
sns.kdeplot(Movies['duration'], shade=True)
```

```python
values_categories =  Movies['listed_in'].value_counts()[:10].tolist()
labels_categories = Movies['listed_in'].value_counts()[:10].index.tolist()
```

```python
plt.figure(figsize=(10,10))
top_categories = Movies['listed_in']
top2 = top_categories.value_counts().head()
plt.pie( values_categories , labels=labels_categories , autopct='%1.0f%%' ) ; 
```

```python
Movies["year_added"].value_counts()
```

```python
plt.figure(figsize=(10 , 6 ))
sns.countplot(x = Movies.year_added)
plt.xticks(rotation = 90);
```

```python
Movies["month_added"].value_counts()
```

```python
sns.countplot(x = Movies.month_added, order = Movies.month_added.value_counts().index)
```

```python
Movies["Day_Name"].value_counts()
```

```python
g = sns.FacetGrid(Movies, col="country", col_order=["United States", "Egypt"], height=5, aspect=1)
g.map(sns.histplot, "Day_Name", stat="percent", common_norm=False)
for ax in g.axes.flat:
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
```

```python
Movies.to_csv("Movies")
```
